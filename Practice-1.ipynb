```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
documents=[
    "The sun rises in the east and sets in the west.",
    "Reading books improves knowledge and vocabulary.",
    "Healthy food and regular exercise keep the body fit."
]
```

```python
embeddings = model.encode(documents)
```

```python
embeddings.shape
```

```python
similarity_matrix = cosine_similarity(embeddings)
```

```python
print(similarity_matrix)
```

```python
query = "sun rises and reading books"
query_embedding =model.encode([query])
scores = cosine_similarity(query_embedding, embeddings)
best_match = np.argmax(scores)
print("Query:", query)
print("Most similar document:", documents[best_match])
print("Similarity score", scores[0][best_match])
```
